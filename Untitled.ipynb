```python
def calculate_null_percentage(df):
    # Calculate the number of null values for each column
    null_counts = df.isnull().sum()
    
    # Calculate the percentage of null values for each column
    total_rows = len(df)
    null_percentages = (null_counts / total_rows) * 100
    
    # Create a new DataFrame to store the results
    result_df = pd.DataFrame({
        'Null Count': null_counts,
        'Null Percentage': null_percentages
    })
    
    return result_df
```

```python
calculate_null_percentage(df)
```

```python
def print_columns_missing_info(df):
    # Calculate the percentage of missing values for each column
    total_rows = len(df)
    missing_percentages = (df.isnull().sum() / total_rows) * 100
    
    # Filter columns with less than 5% missing values and not equal to zero
    columns_missing_info = missing_percentages[(missing_percentages < 5) & (missing_percentages > 0)]
    
    # Print the column names with missing information
    if len(columns_missing_info) > 0:
        print("Columns with missing information (< 5% missing values):")
        for column in columns_missing_info.index:
            print(column)
    else:
        print("No columns have missing information.")
```

```python
def count_unique_values(df, column_name, dropna=False):
    # Get unique values and their counts
    unique_values = df[column_name].value_counts(dropna=dropna).reset_index()
    
    # Rename the columns
    unique_values.columns = [column_name, 'Count']
    
    return unique_values
```

```python
def check_column(df, column_name):
    # Get unique values and their counts
    unique_values = df[column_name].value_counts().reset_index()
    
    # Rename the columns
    unique_values.columns = [column_name, 'Count']
    
    # Plot histogram and box and whisker plot
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
    
    # Histogram
    sns.histplot(df[column_name].dropna(), kde=False, ax=axes[0])
    axes[0].set_xlabel(column_name)
    axes[0].set_ylabel('Frequency')
    axes[0].set_title('Distribution of {}'.format(column_name))
    
    # Box and whisker plot
    sns.boxplot(x=column_name, data=df, ax=axes[1])
    axes[1].set_xlabel(column_name)
    axes[1].set_ylabel('Value')
    axes[1].set_title('Box and Whisker Plot of {}'.format(column_name))
    
    plt.tight_layout()
    plt.show()
    
    return unique_values
```

```python
check_column(df, 'behavior_rank')
```

```python
# Create a function for whether or not an individual is 65 years or older as this
# represents a group at higher risk for serious complications from the flu.

def is_65_plus(x):
    if x['age_group'] == '65+ Years':
        return 1
    else:
        return 0

## apply the function to create new column
df['65_plus_age_group'] = df.apply(lambda x: is_65_plus(x), axis=1)

## check counts of unique values in new col and plot distribution
check_column(df, '65_plus_age_group')
```

```python
## define a function to calculate score for high risk of complications
def high_risk(x):
    risk = 0
    if x['65_plus_age_group'] == 1:
        risk += 1
    if x['child_under_6_months'] == 1:
        risk += 1
    if x['chronic_med_condition'] == 1:
        risk += 1
    return risk

## apply the function to create new column
df['at_high_risk_complications'] = df.apply(lambda x: high_risk(x), axis=1)
```

```python
## check counts of unique values in new col and plot distribution
check_column(df, 'at_high_risk_complications')
```
